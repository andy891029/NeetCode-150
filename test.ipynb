```python
import math
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------
# ✨ 核心修正：設定中文字型與負號顯示
# ---------------------------------------------------------
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']  # 步驟 1：使用微軟正黑體
plt.rcParams['axes.unicode_minus'] = False                 # 步驟 2：讓負號能正常顯示


# --- 1. 保留之前的特定計算 ---
x_point = math.pi / 4
result_point = math.sin(2 * x_point)
print(f"特定計算：當 x = {x_point:.4f} 時，sin(2x) 的值是: {result_point:.4f}")
print("-" * 30)


# --- 2. 用 NumPy 產生大量數據來繪圖 ---
x_all = np.linspace(0, 2 * np.pi, 1000)
y_all = np.sin(2 * x_all)


# --- 3. 用 Matplotlib 繪製圖形 ---
plt.figure(figsize=(10, 6))

# 繪製 sin(2x)
plt.plot(x_all, y_all, label='sin(2x)', color='blue', linewidth=2)

# 添加之前的特定計算點 (用紅色圓點表示)
plt.scatter(x_point, result_point, color='red', s=100, zorder=5, label=f'計算點: x={x_point:.2f}')

# 設置標題和軸標籤
plt.title('繪製 $\sin(2x)$ 的函數圖形', fontsize=16)
plt.xlabel('x (弧度)', fontsize=14)
plt.ylabel('$\sin(2x)$', fontsize=14)

# 顯示網格與參考線
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black', linewidth=0.5)

# 顯示圖例
plt.legend(fontsize=12)

# 顯示圖形
plt.show()
```
